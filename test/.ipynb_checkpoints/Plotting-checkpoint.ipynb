# Computations for the _Dyson log-gas_

```julia
# Value of s (here s = 0)
s = 0.;
```

```julia
# Number of particles 
N = 20;
```

```julia
# Inverse temperature
β = 2;
```

```julia
# Number of bins
nbins = div(N, 2) * 10;
```

```julia
# Creation of the histogram to hold g(r)
pch = RWM.PCHist(N, s, β, nbins);
```

```julia
# We build the Hamiltonian (with periodic bc)
H = build_riesz_hamiltonian(s, N);
```

```julia
# We run the Metropolis-Hasting algorithm
n_samples = 1_000_000 # Total number of samples (i.e. configurations)
n_chains = Threads.nthreads() # Number of parallel chains 
pch = RWM.compute_pair_correlation(N, s, H, β, n_samples, n_chains; nbins);
```

```julia
# We compute the structure factor S(k)
S = RWM.compute_fourier_histogram(pch);
```

```julia
# We display the pair correlation g(r)
p_g = my_plot("Pair correlation \$g(r)\$ of the log-gas")
rr = [N/(2*nbins)*(i-1/2) for i=1:nbins] # List of r's 
plot!(p_g, rr, pch.g, label = "\$g(r)\$", lw = 2, c =:red)

# We display the structure factor S(k)
p_S = my_plot("Structure factor \$S(k)\$ of the log-gas")
kk = 0:1/N:1.5 # List of k's
plot!(p_S, kk, S.(kk), label = "\$S(k)\$", lw = 2, c = :red)

# Full plot 
plot(p_g, p_S, size = (600, 300))
```

#### Reproducing plots from article

```julia
# Path were are located datas
path = "../datas/log_gas";

# We upload the histograms
pch = RWM.load_histogram("$(path)/histograms");

# We upload parameters used 
params = JSON.parsefile("$(path)/parameters.json");

# List of (inverse) temperatures
betas = params["Inverse temperatures"];
nbetas = length(betas);

# Number of particles (here, N=100)
N = pch[1].N;

# Value of s (here, s=0)
s = pch[1].s;
```

##### Plotting the pair correlations $g(r)$ for the _Dyson log-gas_ (Fig. 5)

```julia
# Title (in LaTeX)
title = L"\textrm{Pair\ correlation\ of\ Dyson\ log-gas}"

# Colour palette
cs = palette(:turbo, nbetas)

# Creating the plot
p_g = my_plot(title)

# Labelling the axis
plot!(p_g, xaxis = "\$r\$", yaxis = "\$g(r)\$", framestyle =:box)

# Position of the legend 
plot!(p_g, legend_position = :topright, legendfontsize = 8)

# Maximum distance to be displayed
r_max = 7.

# We display the pair correlations 
for j in eachindex(betas)
    # Retrieving datas
    beta = betas[j]
    nbins = length(pch[j].g)
    
    rs = [N/(2*nbins)*(i-1/2) for i=1:nbins]
    
    plot!(p_g, rs, pch[j].g, c = cs[j],
    lw = 2,
    label = "\$\\beta = $beta \$")
    # Limit range for x 
    plot!(p_g, xlims = (0, r_max))
end

display(p_g)
```

##### Plotting the pair correlations $S(k)$ for the _Dyson log-gas_ (Fig. 6)

```julia
# Title (in LaTeX)
title = L"\textrm{Structure\ factor\ of\ Dyson\ log-gas}"

# Creating the plot
p_s = my_plot(title)

# Labelling the axis
plot!(p_s, xaxis = "\$k\$", yaxis = "\$S(k)\$")

# Position of the legend 
plot!(p_s, legend_position = :topleft, framestyle = :box)

# Cutoff distance for k
k_max = 1.5

# Cutoff distance for S(k)
y_max = 3


# We display the structure factos 
for j in eachindex(betas)
    # Retrieving datas
    beta = Int(betas[j])
    S = RWM.compute_fourier_histogram(pch[j])
    ks = 1/N:1/N:k_max
    
    # Thinning procedure near k = 1
    th = (beta == 1 || beta == 2) ? 0 : 2
    thin = 5
    ks = Array(1/N:thin/N:1-th/N)
    ks = vcat(ks, Array(1-(th-1)/N:1/N:1+(th)/N))
    ks = vcat(ks, Array(1+(th + 1)/N:thin/N:k_max))
    
    # Plotting
    plot!(p_s, ks, S.(ks), c = cs[j],
    lw = 2.5,
    markershape = :circle, markersize= 3.5,
    label = "\$\\beta = $beta\$")

    plot!(p_s, ylims = (0, y_max))
end


display(p_s)
```

##### Plotting regression of $S(k)$ near $k = 0$ and $k = 1$ (Fig. 7)

```julia
# Function to regress the slope (linearly) near k = 0
function slope_S(S, N; n_max = 5)
    ks = 1/N:1/N:n_max/N
    ys = S.(ks)
    lr = linregress(ks, ys)
    a, b = LinearRegression.coef(lr)
    return a
end; 

# Computations of the slope coefficient $C$
coefs = zeros(nbetas)
for j in eachindex(betas)
    S = RWM.compute_fourier_histogram(pch[j])
    coefs[j] = slope_S(S, N; n_max = 2)
end
```

```julia
# Title (in LaTeX)
title = L"\textrm{Linear\ regression\ of\ } S(k) \textrm{\ near\ } k = 0"

# Creating the plot
p_r_0 = my_plot(title)

# Labelling the axis
plot!(p_r_0, xaxis = "\$k\$", yaxis = "\$S(k)\$")

# Position of the legend 
plot!(p_r_0, legend_position = :topleft, framestyle = :box)

# Cutoff distance in k
k_max = 0.2

# Cutoff distance in S(k)
y_max =  0.4


# We plot the structure factors
for j in eachindex(betas)
    # Retrieving datas
    beta = Int(betas[j])
    S = RWM.compute_fourier_histogram(pch[j])
    
    ks = 1/N:1/N:k_max

    # Plotting each g(r)
    plot!(p_r_0, ks, S.(ks), c = cs[j],
    lw = 4,
    markershape = :circle, markersize= 4,
    label = "")

    plot!(p_r_0, ylims = (0, y_max))
end

# We add the regression slopes 
for j in eachindex(betas)
    ks = 1/N:0.001:k_max
    C = round(coefs[j], digits = 2)
    β = Int(betas[j])
    true_C = 2/β
    plot!(p_r_0, ks, coefs[j]*ks, c = cs[j],
        lw = 2.4, ls = :dash, label = "\$C = $C \\simeq $true_C = \\frac{2}{\\beta = $β}\$")
end

# Modifying legend size 
plot!(p_r_0, legendfontsize=7)

display(p_r_0)
```

```julia
# Function to regress \alpha near k = 1
function regress_S_near_1(S::Function, N, n_min, n_max)
    function L(θ) 
        local ℓ(k) = θ[1] * (abs(k - 1)^θ[2])
        ks = n_min/N:1/N:n_max/N
        ys = S.(ks)
        tmp = @. abs2(ℓ(ks) - ys)
        return sum(tmp)/length(ks)
    end
    res = Optim.optimize(L, rand(2))
    C, a = Optim.minimizer(res)
    return C, a
end

# Function to regress on the logarithm near k = 1
function regress_S_near_1_log(S::Function, N, n_min, n_max)
    function L(θ) 
        local ℓ(k) = - θ[1] * log(abs(1-k))
        ks = n_min/N:1/N:n_max/N
        ys = S.(ks)
        tmp = @. abs(ℓ(ks) - ys)
        return sum(tmp)/length(ks)
    end
    res = Optim.optimize(L, [1.]; iterations = 100_000)
    C = Optim.minimizer(res)
    return C
end;
```

```julia
# We compute the \alpha near k = 1
coefs = zeros(2, nbetas)
# Lower and upper bound on the number of points to regress
n_min = N - 4; n_max = N - 1

for j in eachindex(betas)
    S = RWM.compute_fourier_histogram(pch[j].g, N)
    coefs[:, j] .= regress_S_near_1(S, N, n_min, n_max)
end
```

```julia
# Title (in LaTeX)
title = L"\textrm{Regression\ of\ } S(k) \textrm{\ near\ } k = 1"

# Creating the plot
p_r_1 = my_plot(title)

# Labelling the axis
plot!(p_r_1, xaxis = "\$k\$", yaxis = "\$S(k)\$")

# Position of the legend 
plot!(p_r_1, legend_position = :topleft, framestyle =:box)

# Cutoffs of display for the k's
k_max = 1; k_min = 0.9

# Cutoffs for the S(k)'s
y_min = 0.3; y_max =  5

# We do not regress for small beta (i.e. large temperature)
idx_start = 3

# We display the structure factos 
for j in idx_start:nbetas
    # Retrieving datas
    beta = Int(betas[j])
    S = RWM.compute_fourier_histogram(pch[j].g, N)
    
    ks = 1/N:1/N:k_max

    # Plotting each g(r)
    plot!(p_r_1, ks, S.(ks), c = cs[j],
    lw = 3, linealpha = 0.8,
    markershape = :circle, markersize= 3,
    label = "")

    plot!(p_r_1, ylims = (y_min, y_max), xlims = (k_min, k_max))
end

# We regress on the logarithmic divergence for β = 4

idx_2 = 3;
# We compute the structure factor
S = RWM.compute_fourier_histogram(pch[idx_2].g, N)
# We regress on [n_min/N, n_max/N] on the log.
n_min = N - 15; n_max = N - 1
C_2 = regress_S_near_1_log(S, N, n_min, n_max)
ks = n_min/N:0.001:n_max/N
yy = @. - C_2 * log(abs(1 - ks))
# We display the regression
plot!(p_r_1, ks, yy, c = cs[idx_2],
        lw = 3, ls = :dash, 
        label = L"\textrm{Logarithmic\ } -\ln|1-k|")

# We regress on 1/|k - 1|^\alpha for β > 4
for j in (idx_start+1):nbetas
    # Regression window 
    ks = n_min/N:0.001:n_max/N
    C, a = coefs[:, j]
    α = round(a, digits = 2)
    β = Int(betas[j])
    yy = @. C * abs(1 - ks)^a
    true_alpha = 1 - 4/β
    plot!(p_r_1, ks, yy, c = cs[j],
        lw = 3, ls = :dash, 
        label = "\$\\alpha = $(-α) \\simeq $true_alpha = \\ 1 - \\frac{4}{\\beta = $β}\$")
end

# Adjusting the size of the legend and the range of display
plot!(p_r_1, xlims = (k_min, 1.01), legendfontsize = 8)
# We add the k = 1 vertical line 
vline!(p_r_1, [1], c = :black, lw = 2, ls = :dash, label = "")
# Final display
display(p_r_1)
```
